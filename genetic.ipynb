```python
import numpy as np
import matplotlib.pyplot as plt
from mealpy.evolutionary_based.GA import BaseGA, EliteSingleGA
from mealpy import BinaryVar
import os

# ---------------------------
# O objetivo da minha ativvidade é resolver problema da mochila com GA
# ---------------------------


pesos = np.array([8, 4, 7, 2, 6, 10, 3, 5, 11, 9])
valores = np.array([3, 6, 16, 9, 7, 8, 5, 11, 13, 14])
capacidade = 30
n_itens = len(pesos)

# ---------------------------
# Esses valores foram fornecidos pela professora
# ---------------------------


# ---------------------------
# Função de fitness para o problema da mochila
# ---------------------------

# A função de fitness calcula o valor total da mochila e penaliza se exceder a capacidade
# Ela deve retornar um valor negativo, pois o estamos minimizando
# O valor total é a soma dos valores dos itens selecionados
# Se o peso total exceder a capacidade, aplica uma penalidade alta que é o 1e6

# A função fitness_mochila recebe uma solução (vetor binário) e retorna o valor total da mochila só que negativo
# A solução é um vetor binário onde 1 indica que o item está incluído na mochila e 0 indica que não está

def fitness_mochila(solution):
    peso_total = np.sum(pesos * solution)
    valor_total = np.sum(valores * solution)
    if peso_total > capacidade:
        return 1e6  
    return -valor_total

# ---------------------------
# Função para rodar só um experimento
# ---------------------------
def run_experiment(model_class, pop_size, n_gen, mutation_rate, crossover_rate, elitism=False, n_runs=5, exp_name="Experimento"):
    best_scores = []
    
    img_dir = "imgs"
    if not os.path.exists(img_dir):
        os.makedirs(img_dir)
        print(f"Pasta '{img_dir}' criada com sucesso.")

    for run in range(n_runs):
        print(f"{exp_name} - Execução {run+1}")

        problem = {
            "obj_func": fitness_mochila,
            "bounds": BinaryVar(n_vars=n_itens, name="item"),
            "minmax": "min",
        }

        if elitism:
            model = EliteSingleGA(
                epoch=n_gen,
                pop_size=pop_size,
                pc=crossover_rate,
                pm=mutation_rate,
                elite_best=0.05  # 5% da população sendo mantida como a "elite"
            )
        else:
            model = model_class(
                epoch=n_gen,
                pop_size=pop_size,
                pc=crossover_rate,
                pm=mutation_rate
            )

        g_best = model.solve(problem)

        best_fitness = -g_best.target.fitness 
        best_scores.append(best_fitness)

    # métricas de desempenho
    avg_score = np.mean(best_scores)
    std_score = np.std(best_scores)

    print(f"\nResultados do {exp_name} {'(com elitismo)' if elitism else '(sem elitismo)'}")
    print(f"Média do Melhor Valor Total: {avg_score:.2f}")
    print(f"Desvio Padrão: {std_score:.2f}\n")

    # nosso gráfico
    plt.figure(figsize=(8, 5))
    plt.plot(best_scores, marker='o')
    plt.title(f"{exp_name} {'(com elitismo)' if elitism else '(sem elitismo)'}")
    plt.xlabel("Execução")
    plt.ylabel("Melhor Valor Total")
    plt.grid()
    
    # Salvar o gráfico na pasta imgs
    elite_suffix = "com_elitismo" if elitism else "sem_elitismo"
    exp_name_clean = exp_name.replace(" ", "_").replace("-", "_").lower()
    filename = f"{img_dir}/{exp_name_clean}_{elite_suffix}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Gráfico salvo como: {filename}")
    
    plt.show()

# ---------------------------
# config dos experimentos
# ---------------------------
experimentos = {
    "Experimento 1 - Exploração Inicial": {"pop_size": 200, "n_gen": 100, "mutation_rate": 0.05, "crossover_rate": 0.7},
    "Experimento 2 - Exploração e Intensificação Equilibrada": {"pop_size": 150, "n_gen": 75, "mutation_rate": 0.02, "crossover_rate": 0.85},
    "Experimento 3 - Intensificação Focada": {"pop_size": 300, "n_gen": 150, "mutation_rate": 0.1, "crossover_rate": 0.9},
}

# ---------------------------
# aqui é para rodarmos os experimentos, da para ser em um .py separado
# mas como não tem muito código, deixei aqui mesmo
# ---------------------------
if __name__ == "__main__":
    print("\n--- Experimentos sem Elitismo ---\n")
    for nome, config in experimentos.items():
        run_experiment(
            model_class=BaseGA,
            pop_size=config["pop_size"],
            n_gen=config["n_gen"],
            mutation_rate=config["mutation_rate"],
            crossover_rate=config["crossover_rate"],
            elitism=False,
            exp_name=nome
        )

    print("\n--- Experimentos com Elitismo (5% da população) ---\n")
    for nome, config in experimentos.items():
        run_experiment(
            model_class=EliteSingleGA,
            pop_size=config["pop_size"],
            n_gen=config["n_gen"],
            mutation_rate=config["mutation_rate"],
            crossover_rate=config["crossover_rate"],
            elitism=True,
            exp_name=nome
        )
```
